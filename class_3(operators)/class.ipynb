# python logical op

```python
table[3]
```

```python
# and (&&) gate
#        1       2        3        4
print(True and True and True and True) # it will False untill all the statments come True
print(True and True and False and True)
print(True and False and False and True)
print(False and False and False and False)

```

```python
# OR (||) gate
#        1       2        3        4
print(True or True or True or True) # it will true if any of the statments come True
print(True or True or False or True)
print(True or False or False or True)
print(False or False or False or False)# it will False untill all the statments come false
```

```python
# not(!) gate
#        1       2        3        4
print( not True ) # it will convert false iinto true and true into false
print( not False)
print( not False)
print( not True)
```
